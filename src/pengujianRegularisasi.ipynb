# Training and Evaluation on Dataset
Load the dataset, preprocess it for classification, and train the FFNN model with appropriate hyperparameters. Evaluate model performance using accuracy metrics.

```python
from activation import Linear, ReLU, Sigmoid, Tanh, Softmax, LeakyReLU, ELU
from loss import MeanSquaredError, BinaryCrossEntropy, CategoricalCrossEntropy
from initialization import ZeroInitialization, UniformInitialization, NormalInitialization, XavierInitialization, HeInitialization
from model import FFNN
```

### Load Dataset & Data Preparation

```python
mnist = fetch_openml('mnist_784', version=1, as_frame=False, parser='auto')
X = mnist.data.astype('float32')
y = mnist.target.astype('int64')

scaler = StandardScaler()
X = scaler.fit_transform(X)

# Label butuh di-encode dengan One Hot
encoder = OneHotEncoder(sparse_output=False)
y_onehot = encoder.fit_transform(y.reshape(-1, 1))

X_train, X_test, y_train, y_test = train_test_split(X, y_onehot, test_size=0.2, random_state=42)
```

```python
layer_sizes = [784, 128, 32, 10]  # Fitur input ada 784

activations = [
    ELU(),
    ELU(), 
    Softmax()     
]

loss_function = CategoricalCrossEntropy()

initializations = [
    HeInitialization(seed=42),  
    XavierInitialization(seed=42),  
    HeInitialization(seed=42)   
]

model_no_reg = FFNN(
    layer_sizes=layer_sizes,
    activations=activations,
    loss=loss_function,
    initializations=initializations
)

history_no_reg = model_no_reg.train(
    x_train=X_train,
    y_train=y_train,
    batch_size=32,
    learning_rate=0.01,
    epochs=20,
    x_y_val=(X_test, y_test),
    verbose=1
)

# Model dengan regularisasi L1
model_l1 = FFNN(
    layer_sizes=layer_sizes,
    activations=activations,
    loss=loss_function,
    initializations=initializations
)

history_l1 = model_l1.train(
    x_train=X_train,
    y_train=y_train,
    batch_size=32,
    learning_rate=0.01,
    epochs=20,
    x_y_val=(X_test, y_test),
    verbose=1,
    l1_lambda=0.001
)

# Model dengan regularisasi L2
model_l2 = FFNN(
    layer_sizes=layer_sizes,
    activations=activations,
    loss=loss_function,
    initializations=initializations
)

history_l2 = model_l2.train(
    x_train=X_train,
    y_train=y_train,
    batch_size=32,
    learning_rate=0.01,
    epochs=20,
    x_y_val=(X_test, y_test),
    verbose=1,
    l2_lambda=0.001
)


# plot loss tanpa regularisasi
plt.figure(figsize=(10, 6))
plt.plot(history_no_reg['train_loss'], label='Training Loss')
plt.plot(history_no_reg['val_loss'], label='Validation Loss')
plt.title('Train and Val Loss without Regularization')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# prediksi
y_pred = model_no_reg.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(y_test, axis=1)
accuracy = np.mean(y_pred_classes == y_true_classes)
print(f"Test accuracy without Regularization: {accuracy:.4f}")


# plot loss dengan regularisasi L1
# plot loss
plt.figure(figsize=(10, 6))
plt.plot(history_l1['train_loss'], label='Training Loss')
plt.plot(history_l1['val_loss'], label='Validation Loss')
plt.title('Train and Val Loss With L1 Regularization')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# prediksi
y_pred = model_l1.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(y_test, axis=1)
accuracy = np.mean(y_pred_classes == y_true_classes)
print(f"Test accuracy with L1: {accuracy:.4f}")

# plot loss dengan regularisasi L2
plt.figure(figsize=(10, 6))
plt.plot(history_l2['train_loss'], label='Training Loss')
plt.plot(history_l2['val_loss'], label='Validation Loss')
plt.title('Train and Val Loss With L2 Regularization')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# prediksi
y_pred = model_l2.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(y_test, axis=1)
accuracy = np.mean(y_pred_classes == y_true_classes)
print(f"Test accuracy with L2: {accuracy:.4f}")
```

Demonstrate saving the trained model to disk and loading it back, then verify the loaded model produces identical predictions.

```python
# Simpen model ke file
model_no_reg.save("ffnn_model.pkl")

# Load model dari file
loaded_model = FFNN.load("ffnn_model.pkl")

# Cek apakah model yang di-load hasilin prediksi yang sama
y_pred_original = model_no_reg.predict(X_test)
y_pred_loaded = loaded_model.predict(X_test)
identical_predictions = np.allclose(y_pred_original, y_pred_loaded)
print(f"Are predictions identical? {'Yes' if identical_predictions else 'No'}")

# Hitung akurasi untuk model yang di-load
y_pred_loaded_classes = np.argmax(y_pred_loaded, axis=1)
y_true_classes = np.argmax(y_test, axis=1)
accuracy_loaded = np.mean(y_pred_loaded_classes == y_true_classes)
print(f"Test accuracy (loaded model): {accuracy_loaded:.4f}")
```
